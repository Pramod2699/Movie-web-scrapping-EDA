```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
import pandas as pd
```

```python
text1='https://www.themoviedb.org/movie?page='
text2='https://www.themoviedb.org/movie'
```

```python
page_url=[]
for no in range (1,501):
    page_url.append(text1+str(no))
page_link=[]
for link in page_url:
    response=requests.get(link,headers={'User-Agent':'Mozilla/5.0'}).text
    html_data=BeautifulSoup(response,'lxml')
    data_find=html_data.find_all('div',class_='wrapper')
    for i in data_find:
        serial=i.a['href'].replace('/movie','')
        page_link.append(text2+str(serial))
```

```python
len(page_link)
```

```python
all_movies_data=[]

number=0
while number <= 10000:
    resp=requests.get(page_link[number],headers={'User-Agent':'Mozilla/5.0'}).text
    sup=BeautifulSoup(resp,'lxml')
    
    mname=sup.find('h2').text.strip()
    mname=mname.split('\n')[0]
    
    ratings=str(sup.find('span',class_='icon'))
    ratings=ratings.split("icon-r")[1][:2]
    
    genre=sup.find('span',class_='genres').text
    genre=genre.strip()
    
    relase=sup.find('span',class_='release').text
    relase=relase.strip()
    
    run=sup.find('span',class_='runtime')
    run=str(run).replace('<span class="runtime">','')
    run=run.replace('</span>','')
    run=run.strip()
    
    director=sup.find_all('li',class_='profile')
    lst=[]
    dict_name=[]
    for char in director:
        name=char.p.text
        char=char.text.strip().replace('\n',' ')
        lst=char.split(' ')
        for dct in lst:
            if (dct=='Director,' or dct=='Director'):
                dict_name.append(name)
            
    movie_data_dict={
        'name':mname,
        'rating':ratings,
        'genre':genre,
        'release date':relase,
        'runtime':run,
        'director':','.join(dict_name),
        'url':page_link[number]
    }
    
    all_movies_data.append(movie_data_dict)
    number+=1
```

```python
df=pd.DataFrame(all_movies_data)
```

```python
df.to_csv('Movies data')
```
